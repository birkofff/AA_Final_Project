### Handling missing values <a name="missing"></a>

<p>We notice there's a bunch of column with value zero and parameters like glucose, blood pressure, <br>skin thickness, Insulin and BMI cannot be null. let's see how many missing values we have for each column.</p>

```python
missing = df_diabete.drop(df_diabete[['Outcome']], axis=1)
missing_ = missing.isnull()
```

```python
for column in missing_.columns.values.tolist():
    print(column)
    print (missing_[column].value_counts())
    print("") 
```

It seems like there isn't missing values but we notice there's certains features like glucose, Insulin, Blood Pressure, BMI who have zero as value but they cannot be null so let's replace the zeros by NaN.

#### Let's now replace the null values by NaN

```python
df_diabete[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = df_diabete[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
```

<h3>NOTES</h3><br>
To deal with the missing values we will use the SimpleImputer method to replace the NaN by the mean of each category (positive & negative) for the following features: Glucose, Blood pressure and BMI. 

We noticed that SkinThickness and Insulin have a lt of missing values, since they are very important features for the study. We cannot just drop them, we cannot also drop the rows with missing values so we will be using another Imputer metho which is the KNNImputer that replaces the missing value by the mean of the K nearest neighbors.

```python
imputer =SimpleImputer(missing_values=np.nan,strategy="mean")
imputer.fit(positive[["Glucose","BloodPressure","BMI"]])
imputed_data = imputer.transform(positive [["Glucose","BloodPressure","BMI"]].values)
pd.DataFrame(imputed_data, columns=["Glucose","BloodPressure","BMI"])
```

```python
positive.describe()
```

```python
imputer1 =SimpleImputer(missing_values=np.nan,strategy="mean")
imputer1.fit(negative[["Glucose","BloodPressure","BMI"]])
imputed_data2 = imputer1.transform(negative [["Glucose","BloodPressure","BMI"]].values)
pd.DataFrame(imputed_data2, columns=["Glucose","BloodPressure","BMI"])
```

```python
negative.describe()
```

```python
missing2 = positive[["SkinThickness","Insulin"]]
imputer2 = KNNImputer(n_neighbors=2, weights="uniform")
df = pd.DataFrame(imputer2.fit_transform(missing2)).rename(columns={0: "SkinThickness", 1: "Insulin"})
df.head()
```

```python
df.describe()
```

```python
missing3 = negative[["SkinThickness","Insulin"]]
imputer3 = KNNImputer(n_neighbors=2, weights="uniform")
df1 = pd.DataFrame(imputer3.fit_transform(missing3)).rename(columns={0: "SkinThickness", 1: "Insulin"})
df1.head()
```

```python
df1.describe()
```
